```python
names = ['EURJPY', 'EURCHF', 'CHFJPY', 'USDCHF', 'USDCAD']
actions = ['Open', 'High', 'Low', 'Close']
name = 'EURUSD'
df = pd.read_csv(f'scaled/{name}-2019.csv')
df.columns = ['Date', 'Time', f'{name}-Open', f'{name}-High', f'{name}-Low', f'{name}-Close', 'DateTime']
df = df.set_index('DateTime')
start = time()
for name in names:
    print(f'Import {name}')
    _df = pd.read_csv(f'scaled/{name}-2019.csv')
    _df.drop('Date', axis=1, inplace=True)
    _df.drop('Time', axis=1, inplace=True)
    
    _df.columns = [f'{name}-Open', f'{name}-High', f'{name}-Low', f'{name}-Close', 'DateTime']
    
    df = df.join(_df.set_index('DateTime'))
df.to_csv(f'scaled/Merged_2019.csv', float_format='%g', index=False)
```
